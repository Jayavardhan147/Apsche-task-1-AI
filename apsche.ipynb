```python
import pandas as pd
df1=pd.read_csv("Continent.csv")
df2=pd.read_csv("Customers.csv")
df3=pd.read_csv("Transaction.csv")
print(f"Dataframe 1: \n {df1}")
print(f"Dataframe 2: \n {df2}")
print(f"Dataframe 3: \n {df3}")
```

```python
#merge function is used on the basis of customer_id
merge=(pd.merge(df2,df3,on="customer_id"))
print("Merged Customer and Transaction")
print(merge)
# merge.info()
```

```python
#drop_duplicates() function is used to drop duplicates from merged dataframe
merge.drop_duplicates()
```

```python
merge.dropna()
```

```python
#start date and end date columns are converted to date objects by using .to_datetime
merge['start_date']=pd.to_datetime(merge['start_date'],format='%d-%m-%Y',errors='coerce')
merge['end_date']=pd.to_datetime(merge['end_date'],format='%d-%m-%Y',errors='coerce')
merge["Duration"]=(merge['end_date']-merge['start_date']).dt.days
merge.head(20)
```

```python
merge["Duration"].describe()
```

```python
#each customer has multiple custo,er id so .reset_index() function is used so that overall average duration of is calculated
avg_duration = merge.groupby('customer_id')['Duration'].mean().reset_index()
print(avg_duration)
```

```python
#unique fuction helps in finding unique values in given column
unique=df3['txn_type'].unique()
print(unique)
```

```python
#firstly, customer and transaction dataframes are merged on the basis of "customer_id"
#secondly, continent datframe is merged with firstly merged dataframes
#Then groupby() is used to count the each transaction type with respect to each continent. 
merge=pd.merge(df2,df3,on='customer_id')
merge=pd.merge(merge,df1,on="region_id")
count=merge.groupby(["txn_type","region_id"]).size().unstack()
print(count)
```

```python

#three files are converted to pandas dataframe
import pandas as pd
df4=pd.read_csv("Programmer.csv")
df5=pd.read_csv("Software.csv")
df6=pd.read_csv("Studies.csv")
print(f"Dataframe 4: \n {df4} \n")
print(f"Dataframe 5: \n {df5} \n")
print(f"Dataframe 6: \n {df6}")
```

```python

#mean of selling cost is calculated where condition = pascal packages
pas = df5[df5['DEVELOPIN'] == 'PASCAL']
avg = pas['SCOST'].mean()
print("Average Selling Cost for the packages developed in Pascal: ", avg)
```

```python
#names are printed where course opted = DAP
dap=df6[df6['COURSE']=='DAP']
print(dap['PNAME'])
```

```python
#min function is used for corse fee
low=df6["COURSE FEE"].min()
print(f"Lowest course fee {low}")
```

```python
#if SCOST * SOLD > DCOST then revenue is recoverd 
df5['revenue']=df5['SCOST']*df5['SOLD']
recover=df5[df5["revenue"]>= df5["DCOST"]]
print(recover[['PNAME', 'TITLE', 'DEVELOPIN', 'SCOST', 'DCOST', 'SOLD', 'revenue' ]])
```

```python
#max of DCOST makes costliest software development in Basic
cost= df5[df5['DEVELOPIN'] == 'BASIC']['DCOST'].max()
print(f"Cost of the costliest software development in Basic is {cost}")
```

```python
#5000>COURSE FEE<10000
paid = df6[(df6['COURSE FEE']>=5000) & (df6['COURSE FEE']<=10000) ]
print(f"programmers paid 5000 to 10000 for their course are {len(paid)}")
print(paid["PNAME"],["paid"])
```
